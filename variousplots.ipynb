```python
import seaborn as sns 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
tips=sns.load_dataset("tips")
day=tips.groupby("day")["total_bill"].sum().reset_index()
day
```

```python


x = day['day']

y = day['total_bill']

explode = [0,0,0,0.1]

plt.pie(y,labels=x,explode = explode,autopct='%1.1f%%',colors=['red','blue','green','yellow'],
        wedgeprops = {'edgecolor' : 'black'},shadow=True,startangle=90)
plt.show()
```

```python
plt.scatter(tips["total_bill"],tips["tip"])
plt.xlabel("total_bill")
plt.ylabel("tips")
plt.title("scatterplot",fontsize=27)
plt.show()
```

```python
tips['size'].value_counts()
```

```python
plt.scatter(tips["total_bill"],tips["tip"],s=100,c=tips["size"])
plt.xlabel("total_bill")
plt.ylabel("tips")
plt.title("scatter plot")
cbar=plt.colorbar()
cbar.set_label("size")
plt.show()
```
