```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data=sns.load_dataset("iris")
data
```

```python
#matplot
#line graph

plt.plot(data['sepal_length'])
plt.show()
```

```python
plt.plot(data['sepal_width'])
plt.show()
```

```python
plt.plot(data['petal_length'])
plt.show()
```

```python
plt.plot(data['petal_width'])
plt.show()
```

```python
plt.plot(data['petal_length'])
plt.plot(data['petal_width'])
plt.show()
```

```python
plt.plot(data['petal_length'])
plt.plot(data['petal_width'])
plt.xlabel("indexes")
plt.ylabel("petal counts")
plt.title("iris petal",fontsize= 27)
plt.show()
```

```python
plt.plot(data['petal_length'],label="petal_length")
plt.plot(data['petal_width'],label="petal_width")
plt.xlabel("indexes")
plt.ylabel("petal counts")
plt.title("iris petal",fontsize= 27)
plt.legend()
plt.show()
```

```python
plt.plot(data['petal_length'],label = "petal length",linestyle = "--")
plt.plot(data['petal_width'],label="petal width", marker = "o")
plt.xlabel("indexes")
plt.ylabel("petal counts")
plt.title("iris petal",fontsize= 27)
plt.legend()
plt.show()
```

```python
plt.plot(data['petal_length'],label = "petal length",linestyle = "--",linewidth=3)
plt.plot(data['petal_width'],label="petal width", marker = "o")
plt.xlabel("indexes")
plt.ylabel("petal counts")
plt.title("iris petal",fontsize= 27)
plt.legend()
plt.show()
```
